# <font color=#329da8>Visualizing Distribution</font>

### Histogram

***

Histograms allow you to visualize the distribution (how spread out the values are) of a numeric variable for one or several groups. The values are split into bins, each bin is represented as a bar. The height of the bar represents the frequency of that bin. 

```python
### simple histogram to show the distribution of attack power

sns.histplot(df["Attack"], kde = False)
```

```python
## COMPARING THE DISTRIBUTION OF TWO VARIABLE ##

### plotting two histograms in the same figure

plt.figure(figsize=(15,5)) ## change the figure size

sns.histplot(df['Attack'], color = "red", label = "Attack", kde=True) ## plot #1
sns.histplot(df['Defense'], color ="blue", label = "Defense", kde=True) ## plot #2

plt.legend() ## add a legend 
plt.show() ## show both plots in the same figure 
```

### Density

***

Density plots allow to visualize the distribution of a numeric variable for one or several groups. They are very well adapted for large datasets. 

```python
### simple density plot to show the distribution of attack power

sns.kdeplot(df['Attack'], shade = False, color = "Red")
```

```python
### plotting two density plots in the same figure 

plt.figure(figsize=(15,5)) ## change the figure size

sns.kdeplot(df['Attack'], shade = True, color = "red", label = "Attack")
sns.kdeplot(df['Defense'], shade = True, color = "green", label = "Defense")

plt.legend() ## add a legend 
plt.show() ## show both plots in the same figure 
```

### Boxplot

***

A boxplot summarizes the distribution of a numeric variable for one or several groups. It allows to quickly get the median, quartiles and outliers but also hides the dataset individual data points.

<img src="https://miro.medium.com/max/1400/1*2c21SkzJMf3frPXPAR_gZA.png" 
     width="600" 
     height="700" />

```python
### creating a boxplot to show the distribution of attack

sns.boxplot(x = df["Attack"])
```

```python
### creating a series of boxplots to show the distribution of attack across groups

sns.boxplot(x = df["Stage"], y = df["Attack"])
```

```python
### creating a series of boxplots to show the distribution of attack across groups

plt.figure(figsize=(15,5)) # adjust the size of the output plot

sns.boxplot(x = df["Type 1"], y = df["Attack"])

plt.grid() # add grid lines to plot
```

### Violin 

***

A violin plot allow to visualize the distribution of a numeric variable for one or several groups. Each violin represents a group and the shape represents the density estimate of the variable - the more data points in a specific range, the larger the violin is for that range. The inner bars within each violin represent the same values shown with a box plot. 

* The white dot represents the median
* The thicker bar represents the IQR
* The thinner bar represents the 'extreme line' (the highest and lowest value excluding outliers)

```python
### create a violin plot to show the distribution of attack

sns.violinplot(y = df["Attack"])
```

```python
### create a series of violin plots to show the distribution of attack across groups

sns.violinplot(x = df["Stage"], y = df["Attack"])
```

```python
### create a series of violin plots to show the distribution of attack across groups

plt.figure(figsize=(15,5))

sns.violinplot(x = df["Type 1"], y = df["Attack"])

plt.grid()
```

### Scatterplot

***

A scatter plot displays the relationship between 2 numeric variables. Each data point is represented as a circle. Scatterplots will commonly include a trendline to help make the relationship between the variables clearer.

```python
### create a scatterplot to show the linear relationship between attack and defense

sns.regplot(x = df["Attack"], y = df["Defense"], fit_reg = False)

plt.grid()
```

```python
### create a scatterplot to show the linear relationship between attack and defense; with regression line

plt.figure(figsize=(10,10))

sns.regplot(x = df["Attack"], y = df["Defense"], fit_reg = True)

plt.grid()
```

### Detailed Scatterplot

***

Changing the size, shape, and colors of the scatter can be used to represent a third variable, specifically to show variations between groups. 

```python
### create a scatterplot to show the relationship between attack and defense by stage

sns.lmplot(data = df, x = 'Attack', y = 'Defense', fit_reg = False, hue = "Stage")

plt.grid()
```

### Heatmap

***

A heatmap is a graphical representation of data where each value of a matrix is represented as a color. A heatmap can easily be used to color the values of a correlation matrix. 

```python
# show correlation of stats via heatmap
plt.figure(figsize=(7,7))

# calculate the correlation matrix for the numeric dataset 
corr = df.corr()

# plot a heat map to show the magnitude of the correlation between variables
sns.heatmap(corr, vmin = -1, annot = True)
```

### Barplot

***

A barplot shows the relationship between a numeric and a categoric variable. Each group of the categoric variable is represented as a bar. The size of the bar represents a numeric variable value. 


***

#### A Note about Confidence Intervals (CI)

Sometimes when you see a bar graph, you will see lines extended beyond the top of the bar (or overlapping with the bar), these lines are illustrating the confidence interval of the sample. 

When you are graphing data that comes from a sample (compared to data that represents the entire population), the statistics that you calculate are only <u>estimates</u> of the true population value. You're making an educated guess about what the result will be, but there is some room for error. Confidence intervals are used to show between what range you believe the true value is - with some percent of certainty. The more data you have, the smaller the bars will be. Typically, 95% confidence is a safe estimate. 

The CI bar shows the lowest and highest value you are likely to see if you survey the entire population, this is the range in which the true value is thought to be. When you are using a 95% confidence interval, you are saying that you are 95% confident that the true value exists in this range. 

<LEFT><img src='http://www.vista-survey.com/help/images/Confidence_percentage.gif'></LEFT>

In the example above, you can be 95% certain that the actual percent of people who responded "YES" will be between 41% and 58%. The estimated value shown by the bar is 48.6%.  

***

When you are working with data that represents everyone in a population, you do not need to use CI. When plotting a barplot, you will have the choice to include CI's or leave them out. The pokemon dataset represents an entire population -- all pokemon are represented -- so we do not need to use CI's. 

```python
### creating a simple barplot to show the average attack across each stage

sns.barplot(data = df, x = "Stage", y = "Attack", ci = None)

plt.title('Attack Statistic by Pokemon Stage')
plt.grid()
## plt.minorticks_on() 
```

```python
### creating a simple barplot to show the average attack across each stage

sns.barplot(data = df, x = "Stage", y = "Attack", ci = None)

### change the axis labels

plt.xticks(ticks = [0,1,2], labels = ["Stage 1", "Stage 2", "Stage 3"])

### change the axis labels

plt.xlabel("Pokemon Stage")
plt.ylabel("Average Attack")

### add title and grid lines
plt.title('Attack Statistic by Pokemon Stage')
plt.grid()
```

```python
### create a barplot to show average attack across pokemon type

plt.figure(figsize=(20,5))

sns.barplot(data = df, x = "Type 1", y = "Attack", ci = None)

plt.xticks(rotation = 45)
plt.title('Attack Statistic by Pokemon Type')
```

```python
### create a barplot showing the average attack by pokemon type across stage 

plt.figure(figsize=(20,10))

sns.barplot(data = df, x = "Type 1", y = "Attack", hue = "Stage", ci = None)

plt.xticks(rotation = -45)
plt.title('Attack Statistic by Pokemon Type and Stage')
```

### Count Plot

***

A count plot is similar to a bar plot, but the y-axis shows the frequency of an observation instead of representing a numeric variable. Each entity of the categoric variable is represented as a bar. The size of the bar represents the numeric frequency of the group. 

```python
## creating a simple count plot

sns.countplot(data = df, x = "Stage")
```

```python
### creating a count plot to show the number of pokemon type by stage by type of pokemon 

plt.figure(figsize=(20,10))

sns.countplot(data = df, x = "Type 1", hue = "Stage")

plt.xticks(rotation = -45)
plt.title('Frequency of Pokemon Type across Stage')
```
